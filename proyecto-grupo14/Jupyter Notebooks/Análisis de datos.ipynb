# Análisis de Datos 
En este notebook analizaremos algunos de los datos de nuestros datasets, los cuales ya fueron preprocesados en el archivo `preprocesamiento.ipynb`.

## Promedios por cohorte
A continuación se buscarán distribuciones para los promedios de las distintas cohortes a través de herramientas de visualización. Para esto es importante notar que no se debe considerar el PPA del primer semestre de los estudiantes, ya que no tienen alguno.

Se puede apreciar que todas las distribuciones mostradas previamente siguen una distribución muy similar a la normal, por lo que se gráfican normales para poder comparar.

```python
data_2015[data_2015['PPA'] != 0]['PPA'].plot.kde()
mu = data_2015[data_2015['PPA'] != 0]['PPA'].mean()
variance = data_2015[data_2015['PPA'] != 0]['PPA'].var()
sigma = math.sqrt(variance)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))

plt.show()
```

```python
data_2016[data_2016['PPA'] != 0]['PPA'].plot.kde()
mu = data_2016[data_2016['PPA'] != 0]['PPA'].mean()
variance = data_2016[data_2016['PPA'] != 0]['PPA'].var()
sigma = math.sqrt(variance)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))

plt.show()
```

```python
data_2017[data_2017['PPA'] != 0]['PPA'].plot.kde()
mu = data_2017[data_2017['PPA'] != 0]['PPA'].mean()
variance = data_2017[data_2017['PPA'] != 0]['PPA'].var()
sigma = math.sqrt(variance)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))

plt.show()
```

Como se puede ver en los gŕaficos anteriores, la distribución normal se asemeja mucho a las curvas generada por el método `plot` de pandas.

# Distribuciones de ramos del plan común
A continuación se graficarán las distribuciones de notas para los distintos ramos del plan común por semestre. Para esto es necesario juntar las 3 cohortes, ya que queremos obtener la mayor cantidad de datos que podamos por semestre.

```python
data = data_2015.append(data_2016.append(data_2017))
len(data)
```

Se crean 4 funciones para visualizar los datos:
- `plot_distribucion_ramo(sigla, año, semestre)`: Retorna el gráfico de la distribución para un ramo, en un año y semestre especificado.
- `set_color(x)`: Función auxiliar que sirve para visualizar aquellos alumnos que reprobaron o aprobaron el ramo.
- `set_color_diferencia(x)`: Función auxiliar que sirve para visualizar aquellos alumnos a los que el ramo le bajó su ppa vs a aquellos que se los subió.
- `scatter_plot_ramo(sigla, año, semestre)`: Retorna un scatter plot donde en el eje x se encuentra el atributo PPA y en el eje y la nota final del alumno en un ramo, año y semestre especificado.
- `scatter_plot_ramo(sigla, año, semestre)`: Similar al anterior, la diferencia está en que este gráfico muestra en el eje y la diferencia que hay entre la nota que obtuvo el alumno y el ppa del alumno.

```python
def plot_distribucion_ramo(sigla, año, semestre):
    return data.loc[(data['SIGLA'] == sigla) & (data['AÑO'] == año) & (data['SEMESTRE'] == semestre)]['NOTA FINAL'].plot.kde()
```

```python
def set_color(x):
    if x['APROBADO']:
        return 'blue'
    return 'red'

data['COLOR'] = data.apply(set_color, axis=1)
```

```python
def scatter_plot_ramo(sigla, año, semestre):
    new_data = data.loc[(data['SIGLA'] == sigla) & (data['AÑO'] == año) & (data['SEMESTRE'] == semestre)]
    return new_data.plot.scatter(x='PPA',y='NOTA FINAL',c=new_data['COLOR'])
```

```python
def set_color_diferencia(x):
    if x['DIFERENCIA NOTA CON PPA'] >= 0:
        return 'blue'
    return 'red'
data['DIFERENCIA NOTA CON PPA'] = data['NOTA FINAL'] - data['PPA']
data['COLOR DIFERENCIA'] = data.apply(set_color_diferencia, axis=1)
def scatter_plot_diferencia(sigla, año, semestre):
    new_data = data.loc[(data['SIGLA'] == sigla) & (data['AÑO'] == año) & (data['SEMESTRE'] == semestre)]
    return new_data.plot.scatter(x='PPA',y='DIFERENCIA NOTA CON PPA',c=new_data['COLOR DIFERENCIA'])
```

Es muy importante notar que debido a que las cohortes están divididas por año de ingreso no contamos con todas las notas del ramo en un semestre en particular, por lo que la distribución de notas puede cambiar mucho si se varia el semestre.

Por ejemplo si se analizan las notas del curso EYP1113 en el semestre 2016-2 versus al semestre 2017-2 podríamos ver una diferencia, ya que para el semestre 2016-2 sólo tenemos las notas de aquellos alumnos que ingresaron el 2015 y no se han atrasado con respecto a este ramo vs el semestre 2017-2 donde tenemos las notas de alumnos que ingresaron el 2015 y de alumnos que ingresaron el 2016, lo que claramente entregará más información y podría cambiar las distribuciones.

```python
plot_distribucion_ramo('EYP1113', 2016, 2)
plot_distribucion_ramo('EYP1113', 2017, 2)
plot_distribucion_ramo('EYP1113', 2018, 2)
```

Un análisis similar aplica para el semestre 2016-2 versus el semestre 2017-1, la causa de esto puede ser que hay muchas menos notas de alumnos que dieron el ramo el semestre 2017-1 que alumnos que lo dieron el 2016-2.

```python
plot_distribucion_ramo('EYP1113', 2016, 2)
plot_distribucion_ramo('EYP1113', 2017, 1)
print(len(data.loc[(data['SIGLA'] == 'EYP1113') & (data['AÑO'] == 2017) & (data['SEMESTRE'] == 1)]))
print(len(data.loc[(data['SIGLA'] == 'EYP1113') & (data['AÑO'] == 2016) & (data['SEMESTRE'] == 2)]))
```
