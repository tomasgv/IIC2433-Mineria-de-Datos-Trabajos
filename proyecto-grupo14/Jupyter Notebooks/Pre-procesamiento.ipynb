### Cálculo de PPA por semestre
Ahora se calculará el ppa previo al semestre en el que el alumno está cursando los ramos. Es necesario notar que para los novatos esta columna tomará el valor 0, ya que no tiene cursos previos. Sin embargo, esto no afecta la predicción que queremos desarrollar, ya que un novato no puede escoger los ramos que va a tomar.

En primer lugar, se crea la columna `NOTA CRÉDITO`, la cual es simplemente la multiplicación de los créditos del curso con la nota de este. Luego, se eliminan aquellos ramos sin créditos, ya que no son el foco de estudio para nuestro proyecto.

Además, se agregará una nueva columna que nos será útil al momento de predecir, la cual será la columna `CRÉDITOS PREVIOS`, la cual contiene el número de créditos tomados hasta el semestre anterior.

```python
data_2015['NOTA CRÉDITO'] = data_2015['NOTA FINAL'] * data_2015['CRÉDITOS CURSO']
data_2015 = data_2015[data_2015['NOTA CRÉDITO'] != 0]
```

```python
data_2016['NOTA CRÉDITO'] = data_2016['NOTA FINAL'] * data_2016['CRÉDITOS CURSO']
data_2016 = data_2016[data_2016['NOTA CRÉDITO'] != 0]
```

```python
data_2017['NOTA CRÉDITO'] = data_2017['NOTA FINAL'] * data_2017['CRÉDITOS CURSO']
data_2017 = data_2017[data_2017['NOTA CRÉDITO'] != 0]
```

Ahora se itera sobre cada una de las filas de los datasets con el fin de calcular los ppas de cada uno de los semestres de un alumno y almacenarlos en las listas `ppa_2015`, `ppa_2016` y `ppa_2017`.

Además, se llenan las listas `creditos_2015`, `creditos_2016` y `creditos_2017`, las cuales contienen el número de créditos tomados hasta el semestre previo al de la fila.

Se agregan también listas para los atributos de promedio máximo y mínimo, y siglas máximas y mínimas .

```python
ppas_2015 = []
creditos_2015 = []
max_promedios_2015 = []
max_siglas_2015 = []
min_promedios_2015 = []
min_siglas_2015 = []

for i, x in data_2015.iterrows():
    df1 = data_2015.loc[(data_2015['COD ALUMNO'] == x['COD ALUMNO']) & ((data_2015['AÑO'] < x['AÑO']) | (data_2015['AÑO'] == x['AÑO']) & (data_2015['SEMESTRE'] < x['SEMESTRE']))]
    if (len(df1) > 0) :
        ppas_2015.append(df1['NOTA CRÉDITO'].sum()/df1['CRÉDITOS CURSO'].sum())
        creditos_2015.append(df1['CRÉDITOS CURSO'].sum())
        max_promedios_2015.append(df1['NOTA FINAL'].max())
        min_promedios_2015.append(df1['NOTA FINAL'].min())
        max_siglas_2015.append(df1.loc[df1['NOTA FINAL'].idxmax()]['SIGLA'])
        min_siglas_2015.append(df1.loc[df1['NOTA FINAL'].idxmin()]['SIGLA'])
    else:
        ppas_2015.append(0)
        creditos_2015.append(0)
        max_promedios_2015.append(0)
        min_promedios_2015.append(0)
        max_siglas_2015.append(None)
        min_siglas_2015.append(None)
```

```python
ppas_2016 = []
creditos_2016 = []
max_promedios_2016 = []
max_siglas_2016 = []
min_promedios_2016 = []
min_siglas_2016 = []
for i, x in data_2016.iterrows():
    df1 = data_2016.loc[(data_2016['COD ALUMNO'] == x['COD ALUMNO']) & ((data_2016['AÑO'] < x['AÑO']) | (data_2016['AÑO'] == x['AÑO']) & (data_2016['SEMESTRE'] < x['SEMESTRE']))]
    if (len(df1) > 0) :
        ppas_2016.append(df1['NOTA CRÉDITO'].sum()/df1['CRÉDITOS CURSO'].sum())
        creditos_2016.append(df1['CRÉDITOS CURSO'].sum())
        max_promedios_2016.append(df1['NOTA FINAL'].max())
        min_promedios_2016.append(df1['NOTA FINAL'].min())
        max_siglas_2016.append(df1.loc[df1['NOTA FINAL'].idxmax()]['SIGLA'])
        min_siglas_2016.append(df1.loc[df1['NOTA FINAL'].idxmin()]['SIGLA'])
    else:
        ppas_2016.append(0)
        creditos_2016.append(0)
        max_promedios_2016.append(0)
        min_promedios_2016.append(0)
        max_siglas_2016.append(None)
        min_siglas_2016.append(None)
```

```python
ppas_2017 = []
creditos_2017 = []
max_promedios_2017 = []
max_siglas_2017 = []
min_promedios_2017 = []
min_siglas_2017 = []
for i, x in data_2017.iterrows():
    df1 = data_2017.loc[(data_2017['COD ALUMNO'] == x['COD ALUMNO']) & ((data_2017['AÑO'] < x['AÑO']) | (data_2017['AÑO'] == x['AÑO']) & (data_2017['SEMESTRE'] < x['SEMESTRE']))]
    if (len(df1) > 0) :
        ppas_2017.append(df1['NOTA CRÉDITO'].sum()/df1['CRÉDITOS CURSO'].sum())
        creditos_2017.append(df1['CRÉDITOS CURSO'].sum())
        max_promedios_2017.append(df1['NOTA FINAL'].max())
        min_promedios_2017.append(df1['NOTA FINAL'].min())
        max_siglas_2017.append(df1.loc[df1['NOTA FINAL'].idxmax()]['SIGLA'])
        min_siglas_2017.append(df1.loc[df1['NOTA FINAL'].idxmin()]['SIGLA'])
    else:
        ppas_2017.append(0)
        creditos_2017.append(0)
        max_promedios_2017.append(0)
        min_promedios_2017.append(0)
        max_siglas_2017.append(None)
        min_siglas_2017.append(None)
```

Una vez creadas las listas mencionadas previamente, se guardan en una nueva columna de nuestros `DataFrames` llamada `PPA`.

```python
data_2015 = data_2015.assign(PPA=pd.Series(ppas_2015).values)
data_2015 = data_2015.assign(CRÈDITOS_PREVIOS=pd.Series(creditos_2015).values)
data_2015 = data_2015.assign(MAX_PROMEDIO=pd.Series(max_promedios_2015).values)
data_2015 = data_2015.assign(MIN_PROMEDIO=pd.Series(min_promedios_2015).values)
data_2015 = data_2015.assign(MAX_SIGLA=pd.Series(max_siglas_2015).values)
data_2015 = data_2015.assign(MIN_SIGLA=pd.Series(min_siglas_2015).values)
```

```python
data_2016 = data_2016.assign(PPA=pd.Series(ppas_2016).values)
data_2016 = data_2016.assign(CRÈDITOS_PREVIOS=pd.Series(creditos_2016).values)
data_2016 = data_2016.assign(MAX_PROMEDIO=pd.Series(max_promedios_2016).values)
data_2016 = data_2016.assign(MIN_PROMEDIO=pd.Series(min_promedios_2016).values)
data_2016 = data_2016.assign(MAX_SIGLA=pd.Series(max_siglas_2016).values)
data_2016 = data_2016.assign(MIN_SIGLA=pd.Series(min_siglas_2016).values)
```

```python
data_2017 = data_2017.assign(PPA=pd.Series(ppas_2017).values)
data_2017 = data_2017.assign(CRÈDITOS_PREVIOS=pd.Series(creditos_2017).values)
data_2017 = data_2017.assign(MAX_PROMEDIO=pd.Series(max_promedios_2017).values)
data_2017 = data_2017.assign(MIN_PROMEDIO=pd.Series(min_promedios_2017).values)
data_2017 = data_2017.assign(MAX_SIGLA=pd.Series(max_siglas_2017).values)
data_2017 = data_2017.assign(MIN_SIGLA=pd.Series(min_siglas_2017).values)
```

### Eliminar datos sucios
Mediante el análisis de los datos nos dimos cuenta que habían alumnos que tenían ppa 0 cuando estaban dando ramos como EYP1113, lo cual nos perjudica el análisis de los datos, por lo que se eliminarán todos aquellos datos.

Cabe destacar que tenían ppa 0 debido a que no existe registro de los ramos que tomaron previamente. Para encontrar los datos sucios se tuvieron que analizar los distintos semestres y años, abajo se puede ver que la forma de eliminar los datos sucios varian según la cohorte. El caso del alumno 351964 fue un caso especial.

```python
eliminados_2015 = set(list(data_2015[(data_2015['PPA'] == 0) & (data_2015['AÑO'] >= 2017)]['COD ALUMNO']))
eliminados_2016 = set(list(data_2016[(data_2016['PPA'] == 0) & (data_2016['AÑO'] >= 2018)]['COD ALUMNO']))
eliminados_2017 = set(list(data_2016[(data_2016['PPA'] == 0) & (data_2016['AÑO'] >= 2019)]['COD ALUMNO']))
eliminados_2016.add(351964)
```

```python
data_2015 = data_2015[~data_2015['COD ALUMNO'].isin(eliminados_2015)]
data_2016 = data_2016[~data_2016['COD ALUMNO'].isin(eliminados_2016)]
data_2017 = data_2017[~data_2017['COD ALUMNO'].isin(eliminados_2017)]
```
